### 1. Programming Challenge: List of squares

Given an integer N, print all the squares of natural numbers not exceeding

### 3. Programming Challenge: Morning running

On the first day the athlete ran X kilometers, and then every day he increased
his mileage by 10% of the previous value.

Given the number X, determine the number of the day on which the athlete’s
mileage will be at least Y kilometers.

Input format:

The program receives numbers X and Y as input.

Output Format:

The program must print one natural number.

```python
X = int(input("Enter an integer: "))
Y = int(input("Enter an integer: "))

distance = X
i = 1
while distance < Y:
    distance *= 1.1
    i += 1

print(i)
```

### 8. Programming Challenge: Number of elements greater than previous

The sequence consists of natural numbers and ends with the number 0.
Determine how many elements of this sequence are greater than the previous
element.

Input format:

Enter a sequence of integers ending with the number 0 (the number 0 itself
is not included in the sequence, but serves as a sign of its end).

Output Format:

Print the answer to the problem.

```python
count = 0

prev = int(input())
if prev == 0:
    print(0)
    exit()

while True:
    current = int(input())
    if current == 0:
        break
    if current > prev:
        count += 1
    prev = current

print(count)
```

### 12. Programming Challenge: Number reversal

Rearrange the digits of the number in reverse order.

Input format:

A single number N is given

Output Format:

It is necessary to print the digits of this number in reverse order.

```python
n = input()

i = len(n) - 1
while i >= 0:
    print(n[i], end="")
    i = i - 1
```

### 13. Programming Challenge: Complex equation*

Solve the equation in integers: (ax+b) / (cx+d) =0

Input format:

Enter 4 numbers: a,b,c,d; c and d are not equal to zero at the same time.

Output Format:

It is necessary to print all solutions if their number is finite, “NO” (without
quotes) if there are no solutions, and “INF” (without quotes) if there are infinitely
many solutions.

```python
a = int(input())
b = int(input())
c = int(input())
d = int(input())

while True:
    if a == 0:
        if b == 0:
            print("INF")
        else:
            print("NO")
        break

    if -b % a != 0:
        print("NO")
        break

    x = -b // a

    if c * x + d == 0:
        print("NO")
        break

    print(x)
    break
```

### 14. Programming Challenge: Row - 1

Given two integers A and B (with A≤B). Print all numbers from A to B
inclusive.

Input format:

Two integers are entered.

Output Format:

Print the answer to the problem.

```python
A = int(input("Enter first integer: "))
B = int(input("Enter another but greater than first: "))

for i in range(A,B+1):
    print(i,end=" ")
```

### 15. Programming Challenge: Row - 2

Given two integers A and B. Print all numbers from A to B inclusive, in
ascending order if A < B, or in descending order otherwise.

Input format:

Two integers are entered.

Output Format:

Print the answer to the problem.

```python
A = int(input("Enter first integer: "))
B = int(input("Enter another but greater than first: "))

if A < B:
    for i in range(A, B + 1):
        print(i, end=" ")
else:
    for i in range(A, B - 1, -1):
        print(i, end=" ")
```

### 16. Programming Challenge: Flags

Write a program that, given a number n from 1 to 9, displays n flags on the
screen. The image of one flag is 4x4 characters in size, and there is also an empty
(of spaces) column between two adjacent flags. It is allowed to display an empty
column after the last flag. Inside each flag, its number must be written - a number
from 1 to n.

Input format:

Enter a natural number.

Output Format:

Print the answer to the problem.

```python
N = int(input("Enter a natural number: "))

for row in range(4):
    for i in range(1, N + 1):
        if row == 0:
            print("+___", end=" ")
        elif row == 1:
            print(f"|{i} /", end=" ")
        elif row == 2:
            print("|__\\", end=" ")
        elif row == 3:
            print("|   ", end=" ")
    print()
```

### 17. Programming Challenge: Ladder

Given a natural number n≤9, print a ladder of n steps, the i-th step consisting
of numbers from 1 to i without spaces.

Input format:

Enter a natural number.

Output Format:

Print the answer to the problem.

```python
N = int(input("Enter a natural number: "))

num = ""
for i in range(1, N+1):
    num += str(i)
    print(num)
```
