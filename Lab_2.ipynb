## 3. Programming task: Maximum of three numbers

Three integers are given. Find the largest of them (the program should print
exactly one integer).

What is the smallest number of comparison operators (>, <, >=, <=) needed
to solve this problem?

Input format: 

Three integers are entered.

Print the answer to the problem.

## 4. Programming Challenge: Leap Year

A natural number is given. You need to determine whether the year with a
given number is a leap year. 

If the year is a leap year, then print YES, otherwise
print NO. 

Recall that according to the Gregorian calendar, a year is a leap year if its
number is a multiple of 4 but not a multiple of 100, or if it is a multiple of 400.

Input format: 

Enter one natural number.

Output Format: 

Print the answer to the problem.

```python
year = int(input("Enter a number: "))

if year % 400 == 0:
    print("It's a Leap Year!")

elif year%4 == 0:
    if year%100 == 0:
        print("Not a leap year.")
    else:
        print("It's a Leap Year!")
else:
    print("Not a leap year.")
```

## 6. Programming Practice: Apartments

## 8. Programming practice exercise: Chocolate
The chocolate has the shape of a rectangle divided into n×m slices. The
chocolate can be broken once in a straight line into two parts. Determine whether it
is possible to break off a piece of chocolate consisting of exactly k pieces in this
way.

Input format: 

The program receives three numbers as input: n, m, k.

Output Format: 

The program should output one of two words: YES or NO.

```python
n = int(input())
m = int(input())
k = int(input())

if (k % n == 0 and k // n < m) or (k % m == 0 and k // m < n):
    print("YES")
else:
    print("NO")
```

## 10. Programming Practice: Quadrants
Given the coordinates of two points on the plane, you need to determine
whether they lie in the same coordinate quadrant or not (all coordinates are different
from zero).

Input format: 

4 numbers are entered: the coordinates of the first point
(x1,y1) and the coordinates of the second point (x2,y2).

Output Format: 

The program should output the word YES if the points are
in the same coordinate quadrant, otherwise output the word NO.

```python
# Brute-Force

x1 = int(input("Enter the coordinate x1: "))
y1 = int(input("Enter the coordinate y1: "))

x2 = int(input("Enter the coordinate x2: "))
y2 = int(input("Enter the coordinate y2: "))

if ((x1 > 0) and (y1 > 0)) and ((x2 > 0) and (y2 > 0)):
    print("YES")
elif ((x1 > 0) and (y1 < 0)) and ((x2 > 0) and (y2 < 0)):
    print("YES")
elif ((x1 < 0) and (y1 < 0)) and ((x2 < 0) and (y2 < 0)):
    print("YES")
elif ((x1 < 0) and (y1 > 0)) and ((x2 < 0) and (y2 > 0)):
    print("YES")    
else:
    print("NO")
```

```python
# Optimised solution

x1 = int(input("Enter the coordinate x1: "))
y1 = int(input("Enter the coordinate y1: "))

x2 = int(input("Enter the coordinate x2: "))
y2 = int(input("Enter the coordinate y2: "))

if (x1 * x2 > 0) and (y1 * y2 > 0):
    print("YES")
else:
    print("NO")
```

The program receives the numbers A1×B1×C1, A2×B2×C2
as input.

Output Format: 

The program should output one of the following lines:
“Boxes are equal” - if the boxes are the same;
“The first box is smaller than the second one” - if the first box can be placed
in the second one;
“The first box is larger than the second one” - if the second box can be placed
in the first one;
“Boxes are incomparable” - in all other cases.

```python
A1 = int(input("Enter the length of the first box: "))
B1 = int(input("Enter the width of the first box: "))
C1 = int(input("Enter the height of the first box: "))

A2 = int(input("Enter the length of the second box: "))
B2 = int(input("Enter the width of the second box: "))
C2 = int(input("Enter the height of the second box: "))

if A1 > B1:
    temp = A1
    A1 = B1
    B1 = temp
if B1 > C1:
    temp = B1
    B1 = C1
    C1 = temp
if A1 > B1:
    temp = A1
    A1 = B1
    B1 = temp

if A2 > B2:
    temp = A2
    A2 = B2
    B2 = temp
if B2 > C2:
    temp = B2
    B2 = C2
    C2 = temp
if A2 > B2:
    temp = A2
    A2 = B2
    B2 = temp

if A1 == A2 and B1 == B2 and C1 == C2:
    print("Boxes are equal")
elif A1 <= A2 and B1 <= B2 and C1 <= C2 and (A1 < A2 or B1 < B2 or C1 < C2):
    print("The first box is smaller than the second one")
elif A1 >= A2 and B1 >= B2 and C1 >= C2 and (A1 > A2 or B1 > B2 or C1 > C2):
    print("The first box is larger than the second one")
else:
    print("Boxes are incomparable")
```

## 16. Programming Practice: Laptop Storage

Laptops packed in boxes were brought to the warehouse, which is shaped
like a rectangular parallelepiped. 
Each box is also shaped like a rectangular
parallelepiped. According to storage rules, boxes with laptops must be placed in a
warehouse subject to the following two conditions:

The sides of the boxes should be parallel to the sides of the warehouse.
When placed in a warehouse, a box can be placed anywhere (subject to the
previous condition), including on another box, but all boxes must be oriented in the
same way (i.e. one box cannot be placed “standing” and the other “lying”)
Write a program that, based on the size of the warehouse and the size of the
box with the laptop, will determine the maximum number of laptops that can be
placed in the warehouse.

Input format: 

The program receives six natural numbers as input. The first
three define the length, height and width of the warehouse. The next three set the
length, height and width of the box with the laptop, respectively.

Output Format: 

The program should display one number - the maximum
number of laptops that can be placed in the warehouse.

```python
# Ware house dimensions
L = int(input("Enter the length of the warehouse: "))
W = int(input("Enter the width of the warehouse: "))
H = int(input("Enter the height of the warehouse: "))
# Laptop Box Dimensions
l = int(input("Enter the length of the laptop: "))
w = int(input("Enter the width of the laptop: "))
h = int(input("Enter the height of the laptop: "))

# Orientation 1
count1 = (L // l) * (H // h) * (W // w)

# Orientation 2
count2 = (L // l) * (H // w) * (W // h)

# Orientation 3
count3 = (L // h) * (H // l) * (W // w)

# Orientation 4
count4 = (L // h) * (H // w) * (W // l)

# Orientation 5
count5 = (L // w) * (H // l) * (W // h)

# Orientation 6
count6 = (L // w) * (H // h) * (W // l)

max_boxes = count1
if count2 > max_boxes:
    max_boxes = count2
if count3 > max_boxes:
    max_boxes = count3
if count4 > max_boxes:
    max_boxes = count4
if count5 > max_boxes:
    max_boxes = count5
if count6 > max_boxes:
    max_boxes = count6

print(max_boxes)
```
