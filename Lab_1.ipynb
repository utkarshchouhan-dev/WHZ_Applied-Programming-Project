### 1. Programming Challenge: Apple Divide-1

N schoolchildren divide K apples equally, the non-divisible remainder
remains in the basket. How many apples will each student get?

Input data:
3
14

Result:
4

```python
N = int(input("Please enter the number of school children: "))
K = int(input("Please enter the number of apples: "))

print(K//N)
```

## 7. Programming Practice Challenge: Maximum of Two* (do not use loops)

Write a program that reads two integers A and B and outputs the largest of
them. Numbers are integers from 1 to 1000.

When solving the problem, only integer arithmetic operations can be used.

You can not use non-linear constructions: branches, loops, functions.

Input Format:

Two numbers are entered.

Output Format:

Output the answer to the problem.

```python
A = int(input("Enter a number: "))
B = int(input("Enter another number: "))

#Booleans can help in avoiding loops and conditionals.

print( A * (A > B) + B * (B > A) + A * (A == B)) 
```

## 8. Programming Practice: Divisibility Test*

In this problem, you need to check whether the number A is divisible by the
number B. 

You can use only arithmetic operations, the use of any kind of branching, functions, etc. forbidden.

Input Format:

Two natural numbers A and B are introduced.

Output Format:

Print "YES" if A is a multiple of B and "NO" otherwise.

```python
A = int(input("Enter a Number: "))
B = int(input("Enter another Number: "))

#Booleans can help in avoiding loops and conditionals.

print("NO" * (A % B != 0) + "YES" * (A % B == 0)) 
```
